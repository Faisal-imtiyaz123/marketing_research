# Research Problem Definition
The central concern of the research problem is to determine whether there is enough intereset in campus community concerning a campus art market, where different art objects would be sold. It is very important to carry out this research before establishing the market to increase the probability of it's success.
Without enough knowldege about people's interest, organisers and artists won't partake in it.

-------

# Relevance of the Study
Doing this study is very relevant for several reasons as follows:

(a) For Customers: The study would help to know about the customer's interest in art, purchasing power, likes and preferences of different individuals and classes of people such as boys, girls, Btech's, Mtech's, Staff etc. This way we can get a firm understanding of the consumer base.

(b) For Artists: Without the relevant data about consumer base likes and dislikes, purchasing power etc. the artists wouldn't know what kind of art to sell. It would help them adjust their product according to the needs and nature of the market.

(c) For Organisers: The organisers would need to make various decisions regarding date and time of the market, the place of it and other things. So having such data would be really helpful, otherwise the whole organisaion problem would be a mess.

# Literature Review

While specific research on on-campus art markets is limited, we can use some foundational principles from some studies. ( Oldenburg, 1999 ) suggests that transforming underutilised spcaes into important centers for social interaction makes the community stonger. ( Solomon, 2020 ) underscores the importance od convenience and social influence which means , in simple terms, that students are more likely to attend events that are easily accessible and align with their time and place choices and which their friends and peers also find interesting.
(Florida , 2002) says that small markets provide emerging artists with early import chances to showcase the r talent, build a customer base and gain experience. Some studies on event management like ( Getz, 2012) highlights the fact that more the event will understand the customer's logistic preferences such as time, location and payment mehtods, the more likely is it to happen over and over again because of its success each time.

We can summarise all this as follows:

Community Building: On-campus events are critical for strengething the community and inrease interaction.

Convenience is Key: Location and timing are primary determinants of attendance for customers.

Economic Opportunity: Student artists face significant barriers to entering traditional markets; campus-based markets can serve as a great start point to gain customers, knowledge and experience .

Recurrence of an Event: To increase the likelyhood of recurrence of an event because it is successful each time, event logistics need to take into account customer preference's with respect to location, time, spending behaviour and what methods of payment they accept.
